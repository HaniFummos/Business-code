```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib.patches import Patch

try:
    plt.style.use('seaborn-v0_8-whitegrid')
except:
    try:
        plt.style.use('seaborn-whitegrid')
    except:
        plt.style.use('default')

children_gen_mean = children_df['Comp_Gen'].mean()
children_hum_mean = children_df['Comp_Hum'].mean()
parents_gen_mean = parents_df['Comp_Gen'].mean()
parents_hum_mean = parents_df['Comp_Hum'].mean()

_, children_p = stats.wilcoxon(children_df['Comp_Gen'], children_df['Comp_Hum'])
_, parents_p = stats.wilcoxon(parents_df['Comp_Gen'], parents_df['Comp_Hum'])

fig, ax = plt.subplots(1, 1, figsize=(16, 8))

fig.suptitle('Factual Comprehension Scores: Generated vs Gold Standard', fontsize=32, y=1.02, fontweight='bold')

colors = {'Gen': '#808080', 'Hum': '#dbb90d'}

x = np.array([0, 1])
width = 0.35
offset = 0.175

ax.bar(x[0] - offset, children_gen_mean, width, color=colors['Gen'], alpha=0.7, edgecolor='black', linewidth=3, label='Generated')
ax.bar(x[0] + offset, children_hum_mean, width, color=colors['Hum'], alpha=0.7, edgecolor='black', linewidth=3, label='Gold')

ax.bar(x[1] - offset, parents_gen_mean, width, color=colors['Gen'], alpha=0.7, edgecolor='black', linewidth=3)
ax.bar(x[1] + offset, parents_hum_mean, width, color=colors['Hum'], alpha=0.7, edgecolor='black', linewidth=3)

ax.set_ylabel('Correct (out of 4)', fontsize=28, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(['Children', 'Parents'], fontsize=28, fontweight='bold')
ax.tick_params(axis='both', which='major', labelsize=26)
ax.set_ylim(1.8, 3.0)

y_pos = max(children_gen_mean, children_hum_mean) + 0.05
is_sig = children_p < 0.05
p_text = f'p = {children_p:.3f}' if children_p >= 0.001 else 'p < 0.001'
sig_symbol = '*' if is_sig else 'ns'
color = 'red' if is_sig else 'black'
ax.text(x[0], y_pos, f'{sig_symbol} {p_text}', 
        ha='center', va='bottom', fontsize=28, color=color, fontweight='bold' if is_sig else 'normal')

y_pos = max(parents_gen_mean, parents_hum_mean) + 0.05
is_sig = parents_p < 0.05
p_text = f'p = {parents_p:.3f}' if parents_p >= 0.001 else 'p < 0.001'
sig_symbol = '*' if is_sig else 'ns'
color = 'red' if is_sig else 'black'
ax.text(x[1], y_pos, f'{sig_symbol} {p_text}', 
        ha='center', va='bottom', fontsize=28, color=color, fontweight='bold' if is_sig else 'normal')

legend_elements = [
    Patch(facecolor=colors['Gen'], alpha=0.7, label='Generated'),
    Patch(facecolor=colors['Hum'], alpha=0.7, label='Gold'),
]
ax.legend(handles=legend_elements, loc='upper right', fontsize=24, framealpha=0.95, edgecolor='black')

plt.tight_layout()
plt.show()
```
